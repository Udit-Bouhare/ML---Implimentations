```python
import pandas as pd 
import numpy as np 
import matplotlib as plt 
import seaborn as sns 
from sklearn.pipeline import Pipeline 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import r2_score 
import warnings 
warnings.filterwarnings('ignore') 
```

```python
X = 6 * np.random.rand(100, 1) - 3 
y = 0.5 * X**2 + 1.5*X + np.random.randn(100, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.25, random_state=42)
```

```python
def poly_regression(degree): 
    poly_feature = PolynomialFeatures(degree=degree, include_bias=True)
    model = LinearRegression() 
    poly_regression = Pipeline([ 
        ("poly_features", poly_feature), 
        ("Model", model)
    ])
    poly_regression.fit(X_train, y_train)
    y_pred = poly_regression.predict(X_test)
    
    sorted_idx = np.argsort(X_test[:, 0])
    plt.scatter(X_train, y_train, color='blue')
    plt.plot(X_test[sorted_idx], y_pred[sorted_idx], "r-", linewidth=2, label='Predictions')
    plt.legend()
    plt.show()
    
    return poly_regression

poly_regression(2)



```
