```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_classification 
```

```python
# let's create the dataset
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(
      X, y, test_size=0.25, random_state=42)
```

```python
# model training 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# now lets check the accuracy  
from  sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
score = accuracy_score(y_test,y_pred)
matrix = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)

print("Accuracy_score", score)
print("Confision_metrix", matrix)
print("Classification_report", report)
```

## **Hyperparameter Tuning & Cross Validation**

```python
Model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
C_vlaues = [100,10,1.0,0.1,0.01]
solver = ['lbfgs','liblinear','newton-cg','newton-cholesk','sag','saga']
```

```python
params = dict(penalty=penalty,C = C_vlaues,solver = solver)
```

```python
## GridSearchCV 
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=Model, param_grid= params, scoring= 'accuracy', cv= cv, n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_ = grid.predict(X_test)
```

```python
# now lets check the accuracy  
from  sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
score = accuracy_score(y_test,y_pred_)
matrix = confusion_matrix(y_test,y_pred_)
report = classification_report(y_test,y_pred_)

print("Accuracy_score", score)
print("Confision_metrix", matrix)
print("Classification_report", report)
```

## **Randomized SearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV 
CV = RandomizedSearchCV(estimator=Model, param_distributions=params, cv=5, scoring="accuracy")
CV.fit(X_train, y_train)
```

```python
CV.best_params_
```

```python
CV.best_score_
```

```python
y_predict = CV.predict(X_test)
```

```python
# now lets check the accuracy  
from  sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
score = accuracy_score(y_test,y_predict)
matrix = confusion_matrix(y_test,y_predict)
report = classification_report(y_test,y_predict)

print("Accuracy_score", score)
print("Confision_metrix", matrix)
print("Classification_report", report)
```
