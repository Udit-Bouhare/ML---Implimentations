# Data Cleaning

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, "Region"] = 1
df.loc[122:, "Region"] = 0
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df = df.drop(index=[122,123,167], axis=0).reset_index(drop=True)
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
# Change the dtypes
columns = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']

df[columns] = df[columns].astype(int)
```

```python
df.info()
```

```python
# Changing other datatypes to float 
Objects = [features for features in df.columns if df[features].dtypes=='O']
Objects
```

```python
for i in Objects: 
    if i != 'Classes': 
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#  converting classes into numbers 
df['Classes'] = df['Classes'].str.strip()
df['Classes'] = df['Classes'].map({'fire' : 1, 'not fire' : 0})
df['Classes'].value_counts()
```

```python
# Save the dataset
df.to_csv("Algerian_forest_fires_Cleaned_dataset.csv", index=False)
```

# Exploratory Data Analysis (EDA)

```python
df.head()
```

```python
# droped day, month, year
df1 = df.drop(columns=['day','month','year'], axis=1)
```

```python
df1.head()
```

```python
# plot density plot all features
# plt.style.use('seaborn')
df1.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
## percentage for a pie Chart 
percentage =df1['Classes'].value_counts(normalize=True)*100
```

```python
classlabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels, autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```

```python
# Correlation
df1.corr()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df1.corr(), annot=True, cmap="magma")
```

```python
# Boxplots
for col in df1.columns:
    if col not in ['Classes','Region']:
        plt.figure(figsize=(6, 4)) 
        sns.boxplot(x = df1[col],palette="magma")
        plt.show
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13,6))
sns.countplot(x = "month",hue='Classes',data=dftemp)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of fires', weight='bold')
plt.title("Fire Analysis of Bejaia Regions", weight='bold')
plt.show()
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13,6))
sns.countplot(x = "month",hue='Classes',data=dftemp)
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of fires', weight='bold')
plt.title("Fire Analysis of Abbes Region", weight='bold')
plt.show()
```
